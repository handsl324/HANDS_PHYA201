```python
#initial conditions
r0 = [-1000,1000] #initial position of electron
v0 = [10e7, 0] #initial velocity of electron

qp = 1.6e-19 #pos charge
qm = -1.6e-19 #neg charge
mass = 9e-28 #mass of electron

k = 1.4e-20 #boltzmann constant

dt = 0.1
N = 100

def columb(rad):
    k = 1.4e-20
    qp = 1.6e-19 
    qm = -1.6e-19
    force = (k*qp*qm)/(rad**2)
    return force



```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Oct 18 20:31:20 2022

@author: lindseyhands
"""
import math
import matplotlib.pyplot as plt

def pyth(a,b):
    c = (a**2 + b**2)**0.5
    return c

qp = 1.6e-19 #pos charge
qm = -1.6e-19 #neg charge
mass = 9e-28 #mass of electron

k = 1.4e-20 #boltzmann constant

rx = -1e-9 #initial x distance between charges
ry = 1e-9 #initial y distance 
r = pyth(rx,ry)

vx = 10e7 #initial x velocity #10e7 m/s
vy = 0 #initial y velocity
theta = math.atan(ry/rx) #angle from x axis to vector between the charges
#print(theta)
#radius = []

def columb(rad):
    force = (k*qp*qm)/(rad**2)
    return force

def newton(force):
    mass = 9e-28
    acc = force/mass
    return acc


time = [0]
vel = [pyth(vx,vy)]
radx = [rx]
rady = [ry]

step = 0.1

t = 0.1
while t<100:
    accel = newton(columb(r))
    #print(accel)
    
    accelx = accel*math.cos(theta)
    accely = accel*math.sin(theta)
    
    vx = vx + step*accelx
    vy = vy + step*accely
    rx = rx + step*vx + 0.5*accelx*step**2 
    ry = ry + step*vy + 0.5*accely*step**2
    print(rx,vx,accelx) #acceleration is too small to change position
    theta = math.atan(ry/rx)
    v = pyth(vx, vy)
    r = pyth(rx, ry)
    #print(rx,ry)
    #print(theta)
    #radius.append(r)
    time.append(t)
    vel.append(v)
    radx.append(rx)
    rady.append(ry)
    t+=step
    
#plt.scatter(time,rady)   
plt.plot(radx,rady)

plt.grid(True)
plt.show()
```
